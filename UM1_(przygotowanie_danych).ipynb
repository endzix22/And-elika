WYGENEROWANIE DANYCH

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}
df_raw = pd.DataFrame(data=data)
df_raw
```

ZAWSZE PRACUJ NA KOPII DANYCH, STĄD:

```python
df=df_raw.copy()
df.info()
```

```python
## teraz lepiej przetworzyc zmienne object na zmienne kategoryczne, aby zmniejszyc np. pamiec
```

```python
# wybieramy zmienne object
for col in ['size','color','gender','bought']:
  df[col]=df[col].astype('category')
```

```python
# wage zmienmy na float 
df['weight']=df['weight'].astype('float')
df.info()
```

```python
#uzyjmy transponowania do powyzszej kolumny
df.describe().T
# % sa to kwantyle rozkladu
```

```python
df.describe(include=['category']).T
```

```python
## chcemy nasz obiekt przygotowac do modelu uczenia maszynowe
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
##zakladamy klasyfikator binarny:
# df['bought']='yes'-->1
#df['bought']='no'-->0
#teraz bedziemy to na dole robic
```

```python
le=LabelEncoder()
le.fit(df['bought'])  #dopasowuje 0 lub 1 
le.transform(df['bought']) #mapowanie i przygotowanie danych do modelu
```

```python
#zlaczenie 2 wczesniejszych polecen w 1 polecenie
le.fit_transform(df['bought']) 
```

```python
le.classes_
#wyswietla jak wyglada to mapowanie
```

```python
#przekazanie DF do modelu:
df['bought']=le.fit_transform(df['bought'])
df
```

```python
#powrot do pierwotnej postaci kolumny np. 'bought'
df['bought']=le.inverse_transform(df['bought'])
df
```

OneHotEncoder

```python

# do macierzy rzadkich, czytaj jak sa same 0 i 1, to ze wzgledu 
#na pamiec, lepiej to tak przeksztalcic aby posiadac jedynie tylko 
#pozycje np. tylko 1
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

Standaryzacja - StandardScaler


*   std() - pandas nieobciążony
*   std() - numpy obciążony




```python
#Standaryzacja- taka postac gdzie skala wartosci nie ma znaczenia,
# ale ma znaczenie ich wariancja (odchylenie od sredniej)
#np. za duza zmienna wplywajaca na mode i dane potrzebne moga zostac utracone 
#przy projketowaniu modelu
```

```python
df
```

```python
#podglad na srednia i odchylenie standardowe w bibliotece pandas
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
#1 sposob
#zmienienie tych wartosci wyzej na ustandaryzowane czyli skupione wokol 0
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
#2 sposob na to co wyzej tylko przy pomocy napisanej funkcji
def standardize(x):
    return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
#3 sposob użycie gotowej funkcji z biblioteki
from sklearn.preprocessing import scale
scale(df['price'])
#jednak te wartosci sie roznia od tych obliczonych wyzej,
#nie przejmujmy sie, jest to kwestia estymatora obciazonego i nieobciazonego wariancji
```

Najczęciej stosowana metoda na standaryzację:

```python
#4sposob 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```
